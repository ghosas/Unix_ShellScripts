```python
#sort a file, find counts and show counts greater than 1
!sort firstname.txt |uniq -c|awk 'Begin {FS=" "}{if ($1 > 1) print $2,$1}'
```

```python
#Alternative approach
#https://blog.jpalardy.com/posts/alternative-to-sort-uniq-c/
!cat animals.txt | awk '{ cnts[$0] += 1 } END { for (v in cnts) print cnts[v], v }'
```
